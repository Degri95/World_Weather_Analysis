```python
# Import dependencies
import pandas as pd
import numpy as np
import requests
import json
from citipy import citipy
import random
import time
import sys
sys.path.append('C:\\Users\\David\\repos\\data_repos\\World_Weather_Analysis')
from config import weather_api_key
```

```python
# Create a set of 2,000 random latitudes and longitudes.
lats = np.random.uniform(-90.000, 90.000, size=1500)
lngs = np.random.uniform(-180.00, 180.00, size=1500)
lat_lngs = zip(lats, lngs)

coordinates = list(lat_lngs)
```

```python
# Retrieving the nearest city for each pair of coordinates in the list.
cities = []
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    if city not in cities:
        cities.append(city)
# see how many cities were appended.
len(cities)
```

```python
# Perform an API call to retreive information for each city.
city_data = []
measurement = "imperial"
url = "http://api.openweathermap.org/data/2.5/weather?q="
# Print the beginning of logging
print("Beginning Data Retrieval     ")
print("-----------------------------")

# Creating counters.
record_count = 1
set_count = 1

for i, city in enumerate(cities):
    # Group cities into sets of 50 to not overlaod the API.
    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 1
        time.sleep(60)
    
    # Creating and endpoint for each city.
    city_url = url + city.replace(" ", "+") + "&appid=" + weather_api_key + "&units=" + measurement

    # Log the URL, record, and set numbers & the city.
    print(f"Processing Record {record_count} of Set {set_count} | {city}")
    
    # Add 1 to the record count
    record_count += 1
    
    # Run an API request for the cities.
    try:
        city_weather = requests.get(city_url).json()
        # Parse out the needed data.
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        city_description = city_weather["weather"][0]["description"]

        # Append the city information into the city_data list.
        city_data.append({"City": city.title(),
                          "Country": city_country,
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Current Description": city_description
                          })
    except:
        print("City not found. Skipping...")
    


# Indicate the Data Loading is complete.
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```

```python
city_data_df = pd.DataFrame(city_data)
city_data_df.head()
```

```python
output = "WeatherPy_Database.csv"
city_data_df.to_csv(output, index_label="City_ID")
```
